```python
import pandas as pd
messages=pd.read_csv('..\SMSSpamCollection.txt',sep='\t',names=['label','message'])
```

```python
messages
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re

ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if word not in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

Creating Bag of Words

```python
#Output Feature
y=messages['label'].map({'ham': 0, 'spam': 1}).values
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(corpus,y,test_size=0.20)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
X_train
```

```python
#independent feature
X_train=cv.fit_transform(X_train).toarray()
X_test=cv.transform(X_test).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000, formatter=dict(float=lambda x: "%.3g" % x))
X_train
```

```python
cv.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect_model=MultinomialNB().fit(X_train,y_train)
```

```python
y_pred=spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy_score(y_test,y_pred)

```

```python
print(classification_report(y_test,y_pred))
```

```python
### Creating The TF-IDF Model
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(corpus,y,test_size=0.20)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tv=TfidfVectorizer(max_features=2500,ngram_range=(1,2))
X_train=tv.fit_transform(X_train).toarray()
X_test=tv.transform(X_test).toarray()
```

```python
X_train
```

```python
tv.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_tfidf_model =MultinomialNB().fit(X_train,y_train)
```

```python
## Prediction
y_pred=spam_tfidf_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
# Check whether a new message is ham or spam
new_message = 'Congratulations! You have won a free ticket. Call now to claim your prize.'

new_corpus = re.sub('[^a-zA-Z]', ' ', new_message)
new_corpus = new_corpus.lower().split()
new_corpus = [ps.stem(word) for word in new_corpus if word not in stopwords.words('english')]
new_corpus = ' '.join(new_corpus)

new_vector = tv.transform([new_corpus]).toarray()
prediction = spam_tfidf_model.predict(new_vector)[0]

if prediction == 1:
    print('Spam')
else:
    print('Ham')
```

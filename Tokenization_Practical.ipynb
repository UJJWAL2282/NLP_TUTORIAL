corpus refers to paragraph.  document refers to line.   

```python
corpus="""Hello World,I am learning NLP.
The term NLP refer's to Natural Language Processing."""
```

```python
print(corpus)
```

```python
#Tokenization
## sentence --> pargraphs
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import sent_tokenize
```

```python
sentlist=sent_tokenize(corpus)
sentlist
```

```python
##Tokenization
##paragraph --> words
## sentence --> words
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sents in sentlist:
    print(word_tokenize(sents))
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus) # yaha 's is splited into ' and s
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer=TreebankWordTokenizer() ## . -> is not treated is sperate word except last full stop
```

```python
tokenizer.tokenize(corpus)
```
